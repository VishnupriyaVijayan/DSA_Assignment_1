5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
numerical_col=[i for i in df.columns if df[i].dtype=='float64']
```

## univariate analysis 

```python
#histogram of numerical variables

for k in numerical_col:
    plt.figure(figsize=(10,4))
    sns.histplot(data=df,x=k,kde=True)
    plt.title(k)
    
```

```python
#boxplot of numerical variables

for k in numerical_col:
    plt.figure(figsize=(10,4))
    sns.boxplot(data=df,x=k)
    plt.title(k)
```

## bivariate analysis 

```python
for k in numerical_col:
    plt.figure(figsize=(10,4))
    sns.boxplot(data=df,y=k, x='Classification')
    plt.title(k)
```

```python
for k in numerical_col:
    plt.figure(figsize=(10,4))
    sns.stripplot(data=df,y=k,x='Classification')
    plt.title(k)
```
